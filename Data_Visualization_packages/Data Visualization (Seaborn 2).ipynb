```python
%matplotlib inline
import  numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings;
warnings.filterwarnings("ignore")
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins
```

```python
ax = plt.axes();
```

# Scatterplot

## Hue

```python
fig, ax = plt.subplots(1,2, figsize = (12,8))
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'bill_length_mm', ax = ax[0], hue = 'sex');
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', ax = ax[1], hue = 'sex');
```

## Color

```python
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', hue = 'sex', palette = 'viridis');
```

```python
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'flipper_length_mm', hue = 'species', palette = 'rocket');
```

## Density of Data

```python
sns.scatterplot(data = penguins, 
                x = 'body_mass_g',
                y = 'flipper_length_mm', 
                hue = 'species', palette = 'rocket',
                alpha = 0.3 );
```

## S parameter

```python
sns.scatterplot(data = penguins, 
                x = 'body_mass_g',
                y = 'flipper_length_mm', 
                hue = 'species', palette = 'flare',
                alpha = 0.3,
                s = 150 );
```

# Distribution Plots

```python
sns.displot(data = penguins, x = 'species')
```

## Bins

```python
sns.displot(data = penguins, x = 'flipper_length_mm');
```

```python
sns.displot(data = penguins, x = 'flipper_length_mm', bins = 30);
```

```python
plt.figure(figsize = (10,5))
sns.displot(data = penguins, x = 'flipper_length_mm', bins = 30, hue = 'sex', kde = True)
plt.show()
```

```python
fig, ax = plt.subplots(1,2, figsize = (10,5))

sns.histplot(data = penguins, x = 'flipper_length_mm', bins = 30, 
            hue = 'sex', 
            ax = ax[0], 
            kde = True)
sns.scatterplot(data = penguins, x = 'body_mass_g', y = 'bill_length_mm', hue = 'sex', ax = ax[1], s = 150, alpha = 0.3, palette = 'crest');
```

# Facet Grid

```python
g = sns.FacetGrid(penguins, col = 'sex', hue = 'species')
g.map(sns.scatterplot, 'bill_length_mm', 'body_mass_g', alpha = 0.4)
g.add_legend();
```

```python
g = sns.FacetGrid(penguins, col = 'species', hue = 'sex',palette = 'rocket')
g.map(sns.kdeplot, 'bill_length_mm', 'body_mass_g', alpha = 0.8, shade = True)
g.add_legend();
```
