```python
%matplotlib inline
import  numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import warnings;
warnings.filterwarnings("ignore")
```

# Seaborn Datasets

```python
sns.get_dataset_names()
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips.describe().T
```

# Scatter Plot

```python
sns.scatterplot(x = 'size', y = "tip", data = tips)
```

```python
sns.regplot(data=tips, x = "size", y = "tip");
```

# Hue

### use hue to categorize the pscatter plot by color to view differences between categories, such as male/female

```python
plt.figure(figsize=(9,5))
sns.scatterplot(data = tips, y = 'tip', x = 'total_bill', hue = 'sex');
```

# Style

```python
plt.figure(figsize=(9,5))
sns.scatterplot(data = tips, y = 'tip', x = 'total_bill', style = 'smoker');
```

```python
plt.figure(figsize=(9,5))
sns.scatterplot(data = tips, y = 'tip', x = 'total_bill', style = 'smoker', hue = 'day');
```

```python
%matplotlib inline
plt.figure(figsize=(6,5))
sns.scatterplot(data=tips, x = "day", y = "tip", hue = "sex", style="day",);
plt.legend(loc = (1,1))
```

```python
sns.countplot(data = tips, x = 'total_bill')
```
