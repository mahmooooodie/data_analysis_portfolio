# ***Function Method and Object Oriented Methods***

### Functional Methods

```python
plt.plot(age,salary_1, lw = 5,  label = 'salary 1')
plt.plot(age,salary_2, lw = 5, label = 'salary 2')

plt.xlabel('age')
plt.ylabel('salary')
plt.title('salary by age')
plt.legend(bbox_to_anchor = (0,0,))
plt.show()
```

### Object Oriented Method

```python
fig, ax = plt.subplots()
ax.plot(age, salary_1, "b")
ax.set_xlabel("age")
ax.set_ylabel("Salary")
ax.set_title("Salary by Age");
```

### Functional Method (many graphs)

```python
plt.subplot(2 ,1 , 1) # 2 rows, 1 column, 1st graph
plt.plot(age, salary_1, 'r')
plt.subplot(2 ,1 , 2)
plt.plot(age, salary_2, 'b')
plt.show()
```

```python
plt.subplot(2,3,1)
plt.plot(age, salary_1, "r")
plt.subplot(2,3,2)
plt.plot(age, salary_1, "r")
plt.subplot(2,3,3)
plt.plot(age, salary_1, "r")
plt.subplot(2,3,4)
plt.plot(age, salary_1, "b")
plt.subplot(2,3,5)
plt.plot(age, salary_1, "b")
plt.subplot(2,3,6)
plt.plot(age, salary_1, "b")
plt.tight_layout();
plt.show() 
```

### Object Oriented Method (many graphs)

```python
fig, ax = plt.subplots(2, 1)
ax[0].plot(age, salary_1, 'r')
ax[1].plot(age, salary_2, 'b');

plt.tight_layout() # prevents graphs from going in each other when there are labels and such
```

```python
%matplotlib inline
fig, ax = plt.subplots(2, 3)
ax[0][0].plot(age, salary_1, "r")
ax[0][1].plot(age, salary_1, "r")
ax[0][2].plot(age, salary_1, "r")
ax[1][0].plot(age, salary_1)
ax[1][1].plot(age, salary_1)
ax[1][2].plot(age, salary_1)
plt.tight_layout()
plt.show()
```

```python
x = np.arange(0,30,0.2)
fig, ax = plt.subplots(figsize=(14,8))

ax.plot(x, x-1, color = 'green', lw = 3 , linestyle = '-.')
ax.plot(x, x-2, color = 'red', lw = 5 , linestyle = '--')
ax.plot(x, x-3, color = 'blue', lw = 1 , ls = ':')
plt.show()
```

```python
x = np.arange(0,10)

fig, ax = plt.subplots(2,2,figsize=(10,8))
plt.suptitle('Different MArkers and Colors', fontsize =20)

ax[0,0].plot(x, 'r', lw = 5)
ax[0,0].set_title('Graph 1')
ax[0,0].set_xlabel('x')
ax[0,0].set_ylabel('y')
ax[0,1].plot(x-2,'b',lw = 2)
ax[0,1].set_title('Graph 2')
ax[0,1].set_xlabel('x', fontsize=15)
ax[0,1].set_ylabel('y')
ax[1,0].plot(x-3,'g',marker = 'o', lw = 0)
ax[1,0].set_title('Graph 3')
ax[1,0].set_xlabel('x')
ax[1,0].set_ylabel('y')
ax[1,1].plot(x**2, 'b')
ax[1,1].set_title('Graph 4')
ax[1,1].set_xlabel('x')
ax[1,1].set_ylabel('y')
plt.tight_layout()
plt.show()
```

## Pie Charts

```python
x = np.array([100,100,100,100])
mylabels = ['A','B','C','D']
plt.figure(figsize = (10,8))
plt.pie(x, labels = mylabels, labeldistance = 0.75, autopct = '%1.2%%');
```

```python
x = np.array([100,100,100,100])
mylabels = ['A','B','C','D']
myexplode = [0.1,0,0,0]
plt.figure(figsize = (10,8))
plt.pie(x, labels = mylabels, explode = myexplode, labeldistance = 0.75, autopct = '%.2f');
```

```python
x = np.array([120,145,25,10,58])
mylabels = ['A','B','C','D', 'E']
myexplode = [0.1,0,0,0,0]
plt.figure(figsize = (5,5))
plt.pie(x, labels = mylabels, explode = myexplode, labeldistance = 0.75, autopct = '%.1f', startangle = 90, shadow = True);
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 8))
color_list= ["gold", "yellowgreen", "lightcoral", "lightskyblue"]
explode_list = [0.5, 0.5, 0.5, 0.5]
y= np.array([25, 35, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]
plt.pie(y, labels = mylabels,
        labeldistance = 0.7,
        autopct = "%.1f",
        startangle = 90,
        shadow = True,
        pctdistance = 0.5,
       colors = color_list,
       explode = explode_list)
plt.legend(loc = (0))
plt.show()
```
